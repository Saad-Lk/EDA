```python
cats=['Months_on_book',
 'Total_Relationship_Count',
 'Months_Inactive_12_mon',
 'Contacts_Count_12_mon',
 'Credit_Limit',
 'Total_Revolving_Bal',
 'Avg_Open_To_Buy',
 'Total_Amt_Chng_Q4_Q1',
 'Total_Trans_Amt',
 'Total_Trans_Ct',
 'Total_Ct_Chng_Q4_Q1',
 'Avg_Utilization_Ratio']
fig , axis = plt.subplots(3,4,figsize=(20,18))
index = 0
for i in range(3):
    for j in range(4):
        ax= sns.histplot(data=data,x=cats[index] , ax=axis[i][j] , hue='Attrition_Flag' , element="poly")
        index += 1
        
```

```python
catego = ['Gender' ,'Education_Level' , 'Marital_Status' , 'Income_Category' , 'Card_Category' , 'Attrition_Flag']               
fig , axis = plt.subplots(2 , 3 , figsize =(20,15))
p=0
for i in range(2) : 
    for j in range(3) :
        ax = sns.countplot(x=catego[p] , data = data , hue = "Attrition_Flag" , ax=axis[i][j]) 
        p+=1
```

```python
numerical_columns =  ['Customer_Age', 'Credit_Limit' ,'Total_Revolving_Bal','Total_Trans_Amt',
                      'Avg_Utilization_Ratio','Months_Inactive_12_mon']
k = 0 
fig , axis = plt.subplots(2,3,figsize=(35,20))
for i in range(2):
    for j in range(3):
        ax=sns.boxplot(x=numerical_columns[k] , data = data , y="Attrition_Flag" , ax=axis[i][j])
        k += 1
```

```python
numeric_columns = ['Customer_Age','Credit_Limit','Months_on_book','Avg_Utilization_Ratio','Avg_Open_To_Buy',
                  'Total_Trans_Amt','Dependent_count','Total_Relationship_Count','Months_Inactive_12_mon',
                  'Contacts_Count_12_mon','Total_Revolving_Bal', 'Total_Amt_Chng_Q4_Q1','Total_Trans_Ct',
                  'Total_Ct_Chng_Q4_Q1']

data_corr = data.loc[:,numeric_columns].corr()
```

```python
plt.figure(figsize=(20,12))
mask = np.triu(np.ones_like(data_corr)) 
sns.heatmap(data_corr,cmap="YlGnBu", annot=True, mask = mask , )
```
